```python
import cv2
import numpy as np

# Web read image via scikit-image; convert to OpenCV's BGR color ordering
img = cv2.cvtColor(cv2.imread('/content/HYang_4_1.png'), cv2.COLOR_RGB2BGR)

# Inverse binary threshold grayscale version of image
img_thr = cv2.threshold(cv2.cvtColor(img, cv2.COLOR_BGR2GRAY), 128, 255, cv2.THRESH_BINARY_INV)[1]

# Count pixels along the x-axis, find peaks
thr_x = 200
x_sum = np.count_nonzero(img_thr, axis=1)
peaks = np.where(x_sum > thr_x)[0]

# Clean peaks
thr_y = 20

temp = np.diff(peaks).squeeze()
idx = np.where(temp > thr_y)[0]
peaks = np.concatenate(([0], peaks[idx+1]), axis=0) + 1

# Save sub-images
for i in np.arange(peaks.shape[0] - 1):
    cv2.imwrite('sub_image_' + str(i) + '.png', img[peaks[i]:peaks[i+1], :])
```

```python
img = cv2.cvtColor(cv2.imread('/content/JESCO.png6'), cv2.COLOR_RGB2BGR)

# Inverse binary threshold grayscale version of image
img_thr = cv2.threshold(cv2.cvtColor(img, cv2.COLOR_BGR2GRAY), 128, 255, cv2.THRESH_BINARY_INV)[1]

# Count pixels along the y-axis, find peaks
thr_y = 255
y_sum = np.count_nonzero(img_thr, axis=0)
y_peaks = np.where(y_sum > thr_y)[0]

# Clean y-axis peaks
thr_x = 50
temp = np.diff(y_peaks).squeeze()
y_idx = np.where(temp > thr_x)[0]
y_peaks = np.concatenate(([0], y_peaks[y_idx+1]), axis=0) + 1

# Count pixels along the x-axis, find peaks
thr_x = 255
x_sum = np.count_nonzero(img_thr, axis=1)
x_peaks = np.where(x_sum > thr_x)[0]

# Clean x-axis peaks
temp = np.diff(x_peaks).squeeze()
x_idx = np.where(temp > thr_x)[0]
x_peaks = np.concatenate(([0], x_peaks[x_idx+1]), axis=0) + 1

# Save sub-images
for i in np.arange(y_peaks.shape[0] - 1):
    for j in np.arange(x_peaks.shape[0] - 1):
        cv2.imwrite('sub_image_y' + str(i) + '_x' + str(j) + '.png',
                    img[x_peaks[j]:x_peaks[j+1], y_peaks[i]:y_peaks[i+1]])
```

```python
import pandas as pd
import json

# read in the data from the CSV file
df = pd.read_csv('/content/table_data_1.csv', skiprows = 2)
df = df.iloc[:15]

# define a function to convert each row to a dictionary
def row_to_dict(row):
    d = {}
    d['Type'] = row['Type']
    d['C'] = row['C']
    d['Si'] = row['Si']
    d['Mn'] = row['Mn']
    d['P'] = row['P']
    d['S'] = row['S']
    d['Cr'] = row['Cr']
    d['Ni'] = row['Ni']
    d['Cu'] = row['Cu']
    d['Ti'] = row['Ti']
    d['Mo'] = row['Mo']
    d['V'] = row['V']
    d['Al'] = row['Al']
    d['B'] = row['B']
    d['Nb'] = row['Nb']
    d['CE'] = row['CE ( pcm )']
    d['EMI'] = row['EMI']
    d['U.T.'] = row['U.T.']
    d['M.P.I'] = row['M.P.I']
    d['Residual magnetis $ 30GS'] = row['Residual magnetis $ 30GS']
    d['Hydrostatic Test 110S'] = row['Hydrostatic Test 110S']
    d['P( Mpa )'] = row['P( Mpa )']
    return d

# create a list of dictionaries, one for each row in the dataframe
rows = []
for i, row in df.iterrows():
    if row['NO'] == 1:
        d1 = row_to_dict(row)
    elif row['NO'] == 2:
        d2 = row_to_dict(row)
    elif row['NO'] == 3:
        d3 = row_to_dict(row)
    elif row['NO'] == 4:
        d4 = row_to_dict(row)
    elif row['Type'] == 'P':
        if 'P' not in d1:
            d1['P'] = []
            d2['P'] = []
            d3['P'] = []
            d4['P'] = []
        d1['P'].append(row_to_dict(row))
        d2['P'].append(row_to_dict(row))
        d3['P'].append(row_to_dict(row))
        d4['P'].append(row_to_dict(row))
    elif row['Type'] == 'H':
        if 'H' not in d1:
            d1['H'] = []
            d2['H'] = []
            d3['H'] = []
            d4['H'] = []
        d1['H'].append(row_to_dict(row))
        d2['H'].append(row_to_dict(row))
        d3['H'].append(row_to_dict(row))
        d4['H'].append(row_to_dict(row))

# create the final hierarchical dictionary
result = {
    '1': d1,
    '2': d2,
    '3': d3,
    '4': d4
}

# convert the dictionary to JSON
json_str = json.dumps(result, indent=4)

# print the JSON string
print(json_str)
```

```python
import pandas as pd
import unicodedata

def remove_non_english(text):
    if isinstance(text, str):
        return ''.join(c for c in unicodedata.normalize('NFKD', text) if ord(c) < 128)
    else:
        return text


df = pd.read_csv('/content/table_data_1.csv', skiprows = 4, header = 0)
df = df.iloc[:15]


# Remove non-English characters from column headers
df.columns = [remove_non_english(col) for col in df.columns]

# Remove non-English characters from rows
df = df.applymap(remove_non_english)

# Save the cleaned CSV file
df.to_csv('cleaned_file.csv', index=False)
```
