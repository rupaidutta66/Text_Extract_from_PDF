```python
import pandas as pd
import numpy as np
import unicodedata
import json
def remove_non_english(text):
    if isinstance(text, str):
        return ''.join(c for c in unicodedata.normalize('NFKD', text) if ord(c) < 128)
    else:
        return text


df = pd.read_csv('/content/table_TPCO_1.csv', skiprows = 4)
df_1 = df.iloc[:13]

df_2 = df.iloc[17:28]


new_header = df_2.iloc[0]
df_2.columns = new_header
df_2 = df_2.iloc[1:]

# Remove non-English characters from column headers
df_1.columns = [remove_non_english(col) for col in df_1.columns]
df_2.columns = [remove_non_english(col) for col in df_2.columns]

# Remove non-English characters from rows
df_1 = df_1.applymap(remove_non_english)
df_2 = df_2.applymap(remove_non_english)

# # Save the cleaned CSV file
df.to_csv('cleaned_file.csv', index=False)
```
