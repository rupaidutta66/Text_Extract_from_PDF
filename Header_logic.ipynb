```python
from paddleocr import PaddleOCR,draw_ocr
from pdf2image import convert_from_path
from PIL import Image, ImageEnhance

from PIL import Image
import pandas as pd
import numpy as np
import json
import glob

import csv
import os
import re
```

```python

def header_from_pdf(pdf_path):
    P_ocr = PaddleOCR(use_angle_cls=True, lang='en')
    images = convert_from_path(pdf_path)
    j = 0
    for i in range(len(images)):
        images[i].save('/content/page'+ str(i+1) +'.jpg', 'JPEG')
        img_path = '/content/page'+ str(i+1) +'.jpg'

        image = Image.open(img_path)
        if i == 0:
            image = image.rotate(-0.53)
            width, height = image.size

            part_1 = (0, 0, width, int(height * 0.252))
            part_2 = (0, int(height * 0.775), width, height)

            part_1_image = image.crop(part_1)
            part_2_image = image.crop(part_2)

            part_1_image.save('/content/part_1_p'+ str(j+1) +'.jpg')
            part_2_image.save('/content/part_2_p'+ str(j+1) +'.jpg')
            j += 1

        else:
            image = image.rotate(0.53)
            width, height = image.size

            part_1 = (0, 0, width, int(height * 0.255))
            part_2 = (0, int(height * 0.775), width, height)

            part_1_image = image.crop(part_1)
            part_2_image = image.crop(part_2)

            part_1_image.save('/content/part_1_p'+ str(j+1) +'.jpg')
            part_2_image.save('/content/part_2_p'+ str(j+1) +'.jpg')

    k = 0
    for filename in glob.glob('part*.jpg'):
        result = P_ocr.ocr(filename)
        output = result[0]
        boxes = [line[0] for line in output]
        texts = [line[1][0] for line in output]
        scores = [line[1][1] for line in output]

        texts_str = '\n'.join([str(b) for b in texts])
        page_number_1 = filename.split('_')[1]
        page_number_2 = filename.split('_')[2].split('.')[0]
        with open(f'data_{page_number_1}_{page_number_2}.txt', 'w') as file:
            file.write(texts_str)
        k += 1

    filename = '/content/data_1_p1.txt'

    with open(filename, 'r') as file:
        text = file.read()

    lines = [line.strip() for line in text.split('\n') if line.strip()]

    data = {
        'Supplier': lines[0],
        'CERTIFICATE': ' '.join(lines[1:4]),
        'Customer': lines[6],
        'PO NO': lines[8],
        'Product Standard & Name': ' '.join([lines[9], lines[15]]),
        'Treatment way &Delivery Condition': lines[10].replace('NORMALIZED90010', 'NORMALIZED 900°C +/- 10°C'),
        'Certificate No': lines[11].replace('C000000975-1/', 'C000000975-1/2')
    }

    output_file = 'output.json'
    with open(output_file, 'w') as file:
      json.dump(data, file)


pdf_file = '/content/Pipe-ANHUI TIANDA OIL-1.pdf'
header_from_pdf(pdf_file)
```
